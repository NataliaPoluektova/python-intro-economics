## Приклад виконання (Варіант 10)
Коментарі показують логіку обчислень та структуру програми.

```python
# Функція для обчислення нової зарплати після підвищення
def calculate_new_salary(current_salary, increase_percent):
    return current_salary * (1 + increase_percent)

# Функція для перевірки, чи зарплата досягла порогу
def check_threshold(salary, threshold):
    if salary >= threshold:
        return "Зарплата відповідає або перевищує поріг"
    else:
        return "Зарплата нижче порогу"

# Введення даних від користувача
threshold = float(input("Введіть поріг зарплати для перевірки: "))
increase_percent = float(input("Введіть відсоток підвищення зарплати (наприклад, 0.12 для 12%): "))

# Список зарплат за 12 місяців
salaries = []
for month in range(1, 13):
    salary = float(input(f"Введіть зарплату за {month}-й місяць: "))
    new_salary = calculate_new_salary(salary, increase_percent)  # використання функції
    salaries.append(new_salary)  # збереження нової зарплати в список
    print(f"Нова зарплата за {month}-й місяць: {new_salary:.2f}")
    print(check_threshold(new_salary, threshold))  # перевірка порогу

# Обчислення середньої зарплати за рік
total_salary = sum(salaries)  # сума всіх зарплат
average_salary = total_salary / len(salaries)  # середня зарплата
print(f"\nСередня зарплата за рік після підвищення: {average_salary:.2f}")
```
