## Приклад виконання
Умова:
Магазин продає товари. Для кожного товару відомі:

ціна продажу,

собівартість,

кількість проданих одиниць.

Потрібно:
Обчислити прибуток по кожному товару.
Вивести сумарний прибуток магазину.
Визначити, який товар приніс найбільший прибуток.

```python
def analyze_profit(prices, costs, amounts):
    profits = []
    for i in range(len(prices)):
        profit = (prices[i] - costs[i]) * amounts[i]
        profits.append(profit)
        print(f"Товар {i+1}: прибуток = {profit} грн")

    total_profit = sum(profits)
    print("\nСумарний прибуток магазину:", total_profit, "грн")

    max_index = profits.index(max(profits))
    print("Найбільший прибуток приніс товар", max_index + 1)

# Приклад
prices = [120, 200, 150]   # ціна продажу
costs = [80, 150, 100]     # собівартість
amounts = [50, 30, 40]     # кількість проданих одиниць

analyze_profit(prices, costs, amounts)
```
